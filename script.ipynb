```python
#Criação de uma função para criação de testes de hipótese
def testes_de_hipotese(coluna1, coluna2):
    #Criação da tabela de contingência para juntar os dois dataset
    tabela_cont = pd.crosstab(coluna1, coluna2)
    #Recebendo as informações a partir do stats da biblioteca scipy
    #chi2 - qui-quadrado
    #p - valor p
    #graus_lib - graus de liberdade
    #df_esperado - dados esperados em comparação com os reais
    chi2, p, graus_lib, df_esperado = stats.chi2_contingency(tabela_cont)

    #Printando os valores
    print(f'Chi-square: {chi2}')
    print(f'Valor p: {p}')
    print(f'Graus de liberdade: {graus_lib}')
    print('Frequências:')
    #Transformando o array de valores esperados em um dataset para melhor visualização
    df_esperado = pd.DataFrame(df_esperado)
    print(df_esperado)
    
    #Analisando se a hipótese nula irá ser rejeitada ou não
    if p < 0.05:
        print(f'Por conta do valor p ({p}), a hipótese nula foi rejeitada')
    else:
        print(f'Por conta do valor p ({p}), a hipótese nula não foi rejeitada')
```

```python
#Testes de hipóteses para testar a relação entre a idade e os casos de diabetes
diabetes_12 = df[df['Diabetes_012'].isin([1, 2])]

# Agora você pode passar as colunas desejadas para a função testes_de_hipotese
testes_de_hipotese(df['Age'], diabetes_12['Diabetes_012'])

#Hipótese nula rejeitada
#Diabetes tem sim relação com a idade
```

```python
#Testando hipótese para relação entre IMC e Diabetes
testes_de_hipotese(df['BMI'], diabetes_12['Diabetes_012'])

#Hipótese nula rejeitada
#IMC tem relação com diabetes
```

```python
#Testando hipótese para relação entre Colesterol alto e Diabetes
testes_de_hipotese(df['HighChol'], diabetes_12['Diabetes_012'])


#Hipótese nula rejeitada
#Colesterol alto tem relação com diabetes
```

```python
#Testando hipótese para relação entre consumo de Vegetais e Diabetes
testes_de_hipotese(df['Veggies'], diabetes_12['Diabetes_012'])

#Hipótese nula não rejeitada
#Ingestão de vegetais uma ou mais vezes por dia
#não está correlacionado com diabetes
```
