```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
df.isnull().any()
```

```python
#Criando boxplot de todas as colunas do dataset para analisar outliers
for i in df.columns:
    plt.figure()
    bp = plt.violinplot(df[i])
    bp
    print(f'Média da coluna {i}: {df[i].mean()}')
    plt.title(f'Violinplot da coluna: {i}')
    plt.ylabel('Valores')
    
#Todos os boxplots correspondem ao esperado e não contêm nenhum valor exorbitante
```

```python
#Determinando colunas que contém valores diferentes de 0 ou 1,
#esperando o retorno apenas das colunas categóricas, o que foi obtido
for col in df.columns:
    if (~df[col].isin([0, 1])).any():
        print(col)
```

```python
df.describe()
```

```python
#Analisando a média e comparando com a mediana das colunas categóricas
print(f"Média: {df['Diabetes_012'].mean()} Mediana: {df['Diabetes_012'].median()}")
print(f"Média: {df['BMI'].mean()} Mediana: {df['BMI'].median()}")
print(f"Média: {df['GenHlth'].mean()} Mediana: {df['GenHlth'].median()}")
print(f"Média: {df['MentHlth'].mean()} Mediana: {df['MentHlth'].median()}")
print(f"Média: {df['PhysHlth'].mean()} Mediana: {df['PhysHlth'].median()}")
print(f"Média: {df['Age'].mean()} Mediana: {df['Age'].median()}")
print(f"Média: {df['Education'].mean()} Mediana: {df['Education'].median()}")
print(f"Média: {df['Income'].mean()} Mediana: {df['Income'].median()}")

#Resultados onde a mediana foi 0 e diferente da média chamaram atenção, necessitando de boxplot separado
```

```python
plt.hist(df['Age'][df['Diabetes_012']>0])
plt.title('Histograma de faixa de idade por caso diabético afirmativo')
plt.xlabel('Faixa de idade')
plt.show()
```

```python
cont = pd.crosstab(df['Smoker'], df['Diabetes_012'])

cont.plot(kind='bar', stacked=True)
plt.title('Diabetes em comparação com ser fumante ou não')
plt.xticks([0.0, 1.0],['Não fumante', 'Fumante'], rotation='horizontal')
plt.xlabel(None)
plt.show()
```

```python
cont = pd.crosstab(df['PhysActivity'], df['Diabetes_012'])

cont.plot(kind='bar', stacked=True)
plt.title('Diabetes em comparação com exercíos físicos')
plt.xticks([0.0, 1.0],['Inativo', 'Ativo'], rotation='horizontal')
plt.xlabel(None)
plt.show()
```
