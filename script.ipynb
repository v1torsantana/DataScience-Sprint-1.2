```python
#CriaÃ§Ã£o de uma funÃ§Ã£o para criaÃ§Ã£o de testes de hipÃ³tese
def testes_de_hipotese(coluna1, coluna2):
    #CriaÃ§Ã£o da tabela de contingÃªncia para juntar os dois dataset
    tabela_cont = pd.crosstab(coluna1, coluna2)
    #Recebendo as informaÃ§Ãµes a partir do stats da biblioteca scipy
    #chi2 - qui-quadrado
    #p - valor p
    #graus_lib - graus de liberdade
    #df_esperado - dados esperados em comparaÃ§Ã£o com os reais
    chi2, p, graus_lib, df_esperado = stats.chi2_contingency(tabela_cont)

    #Printando os valores
    print(f'Chi-square: {chi2}')
    print(f'Valor p: {p}')
    print(f'Graus de liberdade: {graus_lib}')
    print('FrequÃªncias:')
    #Transformando o array de valores esperados em um dataset para melhor visualizaÃ§Ã£o
    df_esperado = pd.DataFrame(df_esperado)
    print(df_esperado)
    
    #Analisando se a hipÃ³tese nula irÃ¡ ser rejeitada ou nÃ£o
    if p < 0.05:
        print(f'Por conta do valor p ({p}), a hipÃ³tese nula foi rejeitada')
    else:
        print(f'Por conta do valor p ({p}), a hipÃ³tese nula nÃ£o foi rejeitada')
```

```python
#Testes de hipÃ³teses para testar a relaÃ§Ã£o entre a idade e os casos de diabetes
diabetes_12 = df[df['Diabetes_012'].isin([1, 2])]

# Agora vocÃª pode passar as colunas desejadas para a funÃ§Ã£o testes_de_hipotese
testes_de_hipotese(df['Age'], diabetes_12['Diabetes_012'])

#HipÃ³tese nula rejeitada
#Diabetes tem sim relaÃ§Ã£o com a idade
```

```python
#Testando hipÃ³tese para relaÃ§Ã£o entre IMC e Diabetes
testes_de_hipotese(df['BMI'], diabetes_12['Diabetes_012'])

#HipÃ³tese nula rejeitada
#IMC tem relaÃ§Ã£o com diabetes
```

```python
#Testando hipÃ³tese para relaÃ§Ã£o entre Colesterol alto e Diabetes
testes_de_hipotese(df['HighChol'], diabetes_12['Diabetes_012'])


#HipÃ³tese nula rejeitada
#Colesterol alto tem relaÃ§Ã£o com diabetes
```

```python
#Testando hipÃ³tese para relaÃ§Ã£o entre consumo de Vegetais e Diabetes
testes_de_hipotese(df['Veggies'], diabetes_12['Diabetes_012'])

#HipÃ³tese nula nÃ£o rejeitada
#IngestÃ£o de vegetais uma ou mais vezes por dia
#nÃ£o estÃ¡ correlacionado com diabetes
```
