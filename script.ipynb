```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
df.isnull().any()
```

```python
#Criando boxplot de todas as colunas do dataset para analisar outliers
for i in df.columns:
    plt.figure()
    bp = plt.violinplot(df[i])
    bp
    print(f'MÃ©dia da coluna {i}: {df[i].mean()}')
    plt.title(f'Violinplot da coluna: {i}')
    plt.ylabel('Valores')
    
#Todos os boxplots correspondem ao esperado e nÃ£o contÃªm nenhum valor exorbitante
```

```python
#Determinando colunas que contÃ©m valores diferentes de 0 ou 1,
#esperando o retorno apenas das colunas categÃ³ricas, o que foi obtido
for col in df.columns:
    if (~df[col].isin([0, 1])).any():
        print(col)
```

```python
#Analisando a mÃ©dia e comparando com a mediana das colunas categÃ³ricas
print(f"MÃ©dia: {df['Diabetes_012'].mean()} Mediana: {df['Diabetes_012'].median()}")
print(f"MÃ©dia: {df['BMI'].mean()} Mediana: {df['BMI'].median()}")
print(f"MÃ©dia: {df['GenHlth'].mean()} Mediana: {df['GenHlth'].median()}")
print(f"MÃ©dia: {df['MentHlth'].mean()} Mediana: {df['MentHlth'].median()}")
print(f"MÃ©dia: {df['PhysHlth'].mean()} Mediana: {df['PhysHlth'].median()}")
print(f"MÃ©dia: {df['Age'].mean()} Mediana: {df['Age'].median()}")
print(f"MÃ©dia: {df['Education'].mean()} Mediana: {df['Education'].median()}")
print(f"MÃ©dia: {df['Income'].mean()} Mediana: {df['Income'].median()}")

#Resultados onde a mediana foi 0 e diferente da mÃ©dia chamaram atenÃ§Ã£o, necessitando de boxplot separado
```

```python
cont = pd.crosstab(df['Smoker'], df['Diabetes_012'])

cont.plot(kind='bar', stacked=True)
plt.title('Diabetes em comparaÃ§Ã£o com ser fumante ou nÃ£o')
plt.xticks([0.0, 1.0],['NÃ£o fumante', 'Fumante'], rotation='horizontal')
plt.xlabel(None)
plt.show()
```

```python
cont = pd.crosstab(df['PhysActivity'], df['Diabetes_012'])

cont.plot(kind='bar', stacked=True)
plt.title('Diabetes em comparaÃ§Ã£o com exercÃ­os fÃ­sicos')
plt.xticks([0.0, 1.0],['Inativo', 'Ativo'], rotation='horizontal')
plt.xlabel(None)
plt.show()
```
